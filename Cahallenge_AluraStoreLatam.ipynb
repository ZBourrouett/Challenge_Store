### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
Ingreso_tienda1 = tienda["Precio"].sum()
Ingreso_tienda2 = tienda2["Precio"].sum()
Ingreso_tienda3 = tienda3["Precio"].sum()
Ingreso_tienda4 = tienda4["Precio"].sum()

print(f"El ingreso de la tienda 1 fue:{Ingreso_tienda1}")
print(f"El ingreso de la tienda 2 fue:{Ingreso_tienda2}")
print(f"El ingreso de la tienda 3 fue:{Ingreso_tienda3}")
print(f"El ingreso de la tienda 4 fue:{Ingreso_tienda4}")

Ingresos = [Ingreso_tienda1, Ingreso_tienda2, Ingreso_tienda3, Ingreso_tienda4]
Tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]
```

# 2. Ventas por categoría

```python
categoria_tienda1 = tienda.groupby("Categoría del Producto",as_index=False)["Precio"].sum()
categoria_tienda2 = tienda2.groupby("Categoría del Producto",as_index=False)["Precio"].sum()
categoria_tienda3 = tienda3.groupby("Categoría del Producto",as_index=False)["Precio"].sum()
categoria_tienda4 = tienda4.groupby("Categoría del Producto",as_index=False)["Precio"].sum()
print(categoria_tienda1)
print(categoria_tienda2)
print(categoria_tienda3)
print(categoria_tienda4)
```

```python
cantidad_cat_tienda1 = tienda.groupby("Categoría del Producto",as_index=False)["Precio"].count()
cantidad_cat_tienda1.columns = ["categoría", "cantidad ventas"]
cantidad_cat_tienda2 = tienda2.groupby("Categoría del Producto",as_index=False)["Precio"].count()
cantidad_cat_tienda2.columns = ["categoría", "cantidad ventas"]
cantidad_cat_tienda3 = tienda3.groupby("Categoría del Producto",as_index=False)["Precio"].count()
cantidad_cat_tienda3.columns = ["categoría", "cantidad ventas"]
cantidad_cat_tienda4 = tienda4.groupby("Categoría del Producto",as_index=False)["Precio"].count()
cantidad_cat_tienda4.columns = ["categoría", "cantidad ventas"]
print(cantidad_cat_tienda1)
print(cantidad_cat_tienda2)
print(cantidad_cat_tienda3)
print(cantidad_cat_tienda4)

Cant_Cat = [cantidad_cat_tienda1, cantidad_cat_tienda2, cantidad_cat_tienda3, cantidad_cat_tienda4]
```

# 3. Calificación promedio de la tienda


```python
Cal_T1 = tienda["Calificación"].mean()
Cal_T2 = tienda2["Calificación"].mean()
Cal_T3 = tienda3["Calificación"].mean()
Cal_T4 = tienda4["Calificación"].mean()
print(f"La calificación de la tienda 1 es: {Cal_T1}")
print(f"La calificación de la tienda 2 es: {Cal_T2}")
print(f"La calificación de la tienda 3 es: {Cal_T3}")
print(f"La calificación de la tienda 4 es: {Cal_T4}")

Calificación = [Cal_T1, Cal_T2, Cal_T3, Cal_T4]
```

# 4. Productos más y menos vendidos

```python
def obtener_extremos(Tienda,columna_valor,modo):

  conteo_ventas = Tienda[columna_valor].value_counts()

  if modo == "max":
    valor_extremo = conteo_ventas.max()
  elif modo == "min":
      valor_extremo = conteo_ventas.min()
  else:
    raise ValueError("El parámetro 'modo' debe ser 'max' o 'min'.")

  resultado = conteo_ventas[conteo_ventas == valor_extremo]
  return resultado.reset_index(name="extremo ventas")

MejoresT1 = obtener_extremos(tienda, "Producto",modo = "max")
print(f"Tienda mejores 1:\n {MejoresT1}")
PeoresT1 = obtener_extremos(tienda, "Producto",modo = "min")
print(f"Tienda peores 1:\n{PeoresT1}")

MejoresT2 = obtener_extremos(tienda2, "Producto",modo = "max")
print(f"Tienda mejores 2:\n {MejoresT2}")
PeoresT2 = obtener_extremos(tienda2, "Producto",modo = "min")
print(f"Tienda peores 2:\n{PeoresT2}")

MejoresT3 = obtener_extremos(tienda3, "Producto",modo = "max")
print(f"Tienda mejores 3:\n {MejoresT3}")
PeoresT3 = obtener_extremos(tienda3, "Producto",modo = "min")
print(f"Tienda peores 3:\n{PeoresT3}")

MejoresT4 = obtener_extremos(tienda4, "Producto",modo = "max")
print(f"Tienda mejores 4:\n {MejoresT4}")
PeoresT4 = obtener_extremos(tienda4, "Producto",modo = "min")
print(f"Tienda peores 4:\n{PeoresT4}")
```

# 5. Envío promedio por tienda

```python
Env_T1 = tienda["Costo de envío"].mean()
Env_T2 = tienda2["Costo de envío"].mean()
Env_T3 = tienda3["Costo de envío"].mean()
Env_T4 = tienda4["Costo de envío"].mean()
print(f"El costo de envío promedio de la tienda 1 es: {round(Env_T1,2)}")
print(f"El costo de envío promedio de la tienda 2 es: {round(Env_T2,2)}")
print(f"El costo de envío promedio de la tienda 3 es: {round(Env_T3,2)}")
print(f"El costo de envío promedio de la tienda 4 es: {round(Env_T4,2)}")
```

# Gráficos

```python
import matplotlib.pyplot as plt
```

```python
colores = ["skyblue"] * (len(Tiendas) - 1) + ["lightcoral"]
plt.figure(figsize=(10, 6))
plt.bar(Tiendas, Ingresos, color=colores)
plt.xlabel('Ingreso')
plt.ylabel('Tiendas')
plt.title('Ingreso por Tienda')
plt.tight_layout()
plt.show()
```

```python
colores2 = ["skyblue" if calificacion != min(Calificación) else "salmon" for calificacion in Calificación]
plt.figure(figsize=(10, 6))
plt.barh(Tiendas, Calificación, color=colores2)
plt.xlabel('Calificación')
plt.ylabel('Tiendas')
plt.title('Calificación por Tienda')
plt.tight_layout()
```

```python
datos_tiendas = [tienda, tienda2, tienda3, tienda4]


fig, axes = plt.subplots(1, 4, figsize=(20, 5))
fig.suptitle("Distribución de ventas por categoría", fontsize=16, y=1.05)

colores = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#FFBE0B', '#8AC926']

for i, (df, nombre) in enumerate(zip(datos_tiendas, Tiendas)):

    ventas_por_categoria = df.groupby("Categoría del Producto")["Precio"].count()

    # Gráfico
    axes[i].pie(
        ventas_por_categoria,
        labels=ventas_por_categoria.index,
        autopct='%1.1f%%',
        startangle=90,
        colors=colores[:len(ventas_por_categoria)],
        wedgeprops={'edgecolor': 'white', 'linewidth': 0.5},
        textprops={'fontsize': 9}
    )
    axes[i].set_title(nombre, fontsize=12, pad=10)

plt.tight_layout()
plt.show()
```
